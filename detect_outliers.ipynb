```python
import pandas as pd
import numpy as np
data ={
    'Score':[85,88,90,87,92,95,250,89,91,88]
}
df = pd.DataFrame(data)
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 *IQR
upper_bound = Q3 + 1.5 *IQR
detect = df[(df['Score'] >= lower_bound) & (df['Score'] <= upper_bound)]
print("Original data:\n",df)
print("\n Without outliers:\n",detect)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 6))
plt.boxplot(df['Score'], vert=True, patch_artist=True)
plt.title("Boxplot of Scores")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,2)
plt.boxplot(detect['Score'], patch_artist=True)
plt.title("without Outliers")
plt.tight_layout()
plt.show()
```
