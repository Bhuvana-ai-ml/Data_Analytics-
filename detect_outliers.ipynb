```python
import pandas as pd
import numpy as np
data ={
    'Score':[85,88,90,87,92,95,250,89,91,88]
}
df = pd.DataFrame(data)
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 *IQR
upper_bound = Q3 + 1.5 *IQR
detect = df[(df['Score'] >= lower_bound) & (df['Score'] <= upper_bound)]
print("Original data:\n",df)
print("\n Without outliers:\n",detect)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 6))
plt.boxplot(df['Score'], vert=True, patch_artist=True)
plt.title("Boxplot of Scores")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,2)
plt.boxplot(detect['Score'], patch_artist=True)
plt.title("without Outliers")
plt.tight_layout()
plt.show()
```

```python


data = {
    'Hours_Studied': [2, 3, 4, 5, 6, 7, 8, 9, 10, 12],
    'Score': [50, 55, 60, 62, 70, 74, 78, 85, 88, 95],
    'Sleep_Hours': [6, 6.5, 6.5, 7, 7, 7.5, 7.5, 8, 8, 8.5]
}

df = pd.DataFrame(data)

plt.figure(figsize=(7, 5))
plt.scatter(df['Hours_Studied'], df['Score'], color='blue')
plt.title('Scatter Plot: Hours Studied vs Score')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.grid(True)
plt.show()

```

```python

import numpy as np
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(6, 5))
cax = ax.matshow(corr_matrix, cmap='coolwarm')
fig.colorbar(cax)

ax.set_xticks(np.arange(len(corr_matrix.columns)))
ax.set_yticks(np.arange(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='left')
ax.set_yticklabels(corr_matrix.columns)

plt.title('Correlation Heatmap', pad=20)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
